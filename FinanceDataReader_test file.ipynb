```python
samsung = f'YAHOO:005930.KS'

price_df = fdr.DataReader(samsung, '2000-01-01', '2025-02-01')
display(price_df)
```

```python
price_df[['Open', 'High', 'Low', 'Close', 'Adj Close']].plot()
plt.yscale('log', base = 2)
plt.show()
```

```python
# https://grand-unified-engine.tistory.com/17       # 이동평균선이란?
# https://smcstudy.tistory.com/31                   # What is the func. 'rolling' in pandas
df = price_df
df = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
df['MA3'] = df['Close'].rolling(window=3).mean()    # 3일선
df['MA5'] = df['Close'].rolling(window=5).mean()    # 5일선
df['MA10'] = df['Close'].rolling(window=10).mean()  # 10일선

df
```
