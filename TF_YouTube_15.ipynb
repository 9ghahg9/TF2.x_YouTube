## <span style="font-size:25px">CIFAR-10 CNN Example</span>

```python
# CIFAR-10 데이터 불러오기 및 정규화

import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

(x_train, t_train), (x_test, t_test) = cifar10.load_data()

x_train = x_train.reshape(-1, 32, 32, 3)    # chennel이 3인 이유는,
                                            # Red, Green, Blue 이렇게 삼원색으로 
                                            # 이뤄져있기 때문이다.
# https://bruders.tistory.com/85 에서 확인
x_test = x_test.reshape(-1, 32, 32, 3)

print(x_train.shape, x_test.shape)
print(t_train.shape, t_test.shape)

x_train = x_train.astype(np.float32) / 255.0
x_test = x_test.astype(np.float32) / 255.0
```

```python
# 모델 구축

cnn = Sequential()

cnn.add(Conv2D(input_shape=(32, 32, 3), kernel_size=(3, 3), 
               filters=32, activation='relu'))
cnn.add(Conv2D(kernel_size=(3, 3), 
               filters=64, activation='relu'))
cnn.add(MaxPool2D(pool_size=(2, 2)))
cnn.add(Flatten())
cnn.add(Dense(128, activation='relu'))
cnn.add(Dropout(0.5))
cnn.add(Dense(10, activation='softmax'))
```

```python
# 모델 컴파일 및 학습
cnn.compile(loss='sparse_categorical_crossentropy', 
            optimizer=Adam(), 
            metrics=['accuracy'])
hist = cnn.fit(x_train, t_train, batch_size=128, epochs=30, validation_data=(x_test, t_test))
```

```python
# 모델 평가

cnn.evaluate(x_test, t_test)
```

```python
# model trend

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title('Loss Trend')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'validation'], loc='best')
plt.grid()
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])

plt.subplot(1, 2, 2)
plt.title('Accuracy Trend')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'], loc='best')
plt.grid()
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])

plt.show()
```

## <span style="font-size:25px">CNN 성능 향상(오버피팅\, 정확도/)</span>

<dl style="font-size:20px;">
<dt>
▶ 더 많은 레이어(layer) 쌓기
</dt><br>
<dt style="font-size:18px;">
- 컨볼루션 레이어가 중첩된 더 깊은 구조가 될수록 성능은 크게 개선됨
<br><br>
- AlexNet(2012, 8 layers), VGGNet(2014, 19 layers), 
<br>GoogleLeNet(2014, 22 layers), ResNet(2015, 152 layers),...
</dd>
</dl>

<div style="border-top: 3px dashed rgb(83, 197, 87);margin: 20px;width: 30%;"></div>

<dl style="font-size:20px;">
<dt>
▶ 이미지 데이터 보강(Image Data Augmentation)
</dt><br>
<dt style="font-size:18px;">
- 딥러닝에서는 많은 학습 데이터를 사용하면 성능을 개선 시킬 수 있음
<br><br>
- 기존의 데이터(이미지)가 있을 때, 해당 데이터를 원본으로 해서 다양한 변형 
<br>(rotate, shear, zoom, shift, horizontal flip,...)을 주고 이렇게 생선된 데이터를 
<br>원본 학습 데이터에 포함시켜 수 많은 학습 데이터를 확보할 수 있음
</dd>
</dl>

<div style="border-top: 3px dashed rgb(83, 197, 87);margin: 20px;width: 30%;"></div>
